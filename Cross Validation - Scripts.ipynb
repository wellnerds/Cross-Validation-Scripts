**Cross Validation - Entendo de forma simples da validação**

```
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn import datasets
from sklearn import svm
```

**Carregando datast**

```
iris = datasets.load_iris()
```

```
iris.target
```

**Holdout**


*   Separa os dados em treino e teste.
*   Separa aleatoriamente em 70% dos dados para treino e 30% dos dados para teste.



```
# O tamanho dos dados de teste é definido pelo parâmetro test_size

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3)
```

**Instancia o classificador SVM**

```
clf = svm.SVC(gamma='auto')
```

**Treina o algoritmo e gera um modelo**

```
clf.fit(X_train, y_train)
```

**Avalia o modelo com os dados de teste**

```
clf.score(X_test, y_test)
```




*   Usamos a função da scikit-learn cross_val_score() para automatizar esse método.




**Executa Cross Validation com 5 Folds**

```
# a função cross_val_score recebe por parâmetro o modelo, 
# todos os dados para treino, os dados de classes, a quantidade de fodls e o método de avaliação

scores = cross_val_score(clf, iris.data, iris.target, cv=5, scoring='accuracy')
```

**Imprimindo os resultados dos folds**

```
scores
```

**Média dos resultados**

```
scores.mean()
```
